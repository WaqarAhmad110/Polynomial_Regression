# Ploynomial Regression:
Polynomial Regression is a extension of linear regression in which the relationship between the independent variable x and dependent variable y is modelled as an nth-degree polynomial. Polynomial regression fits a nonlinear relationship between the value of x and the corresponding conditional mean of y, denoted E(y | x).
•	There are some relationships that a researcher will hypothesize is curvilinear. Clearly, such types of cases will include a polynomial term.
•	Inspection of residuals. If we try to fit a linear model to curved data, a scatter plot of residuals (Y-axis) on the predictor (X-axis) will have        patches of many positive residuals in the middle. Hence in such a situation, it is not appropriate.   

The general form of a polynomial regression equation of degree n:
 

y=β0+β1x+β2x2+β3x3+…+βnxn+ϵ

•	y: Dependent variable (target variable).
•	x: Independent variable (input feature).
•	β0,β1,β2,…,βn\beta_0, \beta_1, \beta_2, \ldots, \beta_nβ0,β1,β2,…,βn: Coefficients (parameters to be learned during training).
•	nnn: Degree of the polynomial (determines the model complexity).
•	ϵ\epsilonϵ: Error term (accounts for noise in the data).

The basic goal of regression analysis is to model the expected value of a dependent variable y in terms of the value of an independent variable x.


You are tasked with predicting monthly sales based on advertising spending. The data suggests a non-linear relationship between advertising spending and sales.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression


X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1.5, 3.1, 5.0, 7.2, 10.5])

poly = PolynomialFeatures(degree=2) 
X_poly = poly.fit_transform(X)
model = LinearRegression()
model.fit(X_poly, y)

y_pred = model.predict(X_poly)

plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Polynomial Fit')
plt.xlabel("Advertising ($1000)")
plt.ylabel("Sales ($1000)")
plt.legend()
plt.title("Polynomial Regression: Sales vs Advertising")
plt.show()
```

# A bike manufacturer wants to predict the speed of a bike based on the gear ratio. The relationship between the gear ratio and speed is non-linear.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Data
X = np.array([1.0, 1.5, 2.0, 2.5, 3.0]).reshape(-1, 1)
y = np.array([5.0, 7.5, 13.0, 22.5, 30.0])

# Transform to Polynomial Features
poly = PolynomialFeatures(degree=2)  # Quadratic
X_poly = poly.fit_transform(X)

# Fit the Polynomial Regression Model
model = LinearRegression()
model.fit(X_poly, y)

# Predictions
y_pred = model.predict(X_poly)

# Visualization
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Polynomial Fit')
plt.xlabel("Gear Ratio")
plt.ylabel("Speed (km/h)")
plt.title("Polynomial Regression: Speed vs Gear Ratio")
plt.legend()
plt.show()
```
